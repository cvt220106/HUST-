## 显示Euler法求解初值问题
$$
显示Euler法：y_{n+1}=y_n+hf(x_n,y_n) \\
y'=-\frac{0.9}{1+2x}y, y(0)=1,x\in(0,1) 
$$
求解x=0,0.02,0.04,...,0.10时的数值解

```python
import numpy as np
import matplotlib.pyplot as plt
def init_y(x):
    return (1+2*x)**(-0.45)

def error(y):
    return (1+2*x)**(-0.45)-y

x=np.linspace(0,0.1,6)
print(x)
y = [0]*len(x)
y[0]=1
for i in range(1,len(y)):
    y[i] = y[i-1]-0.02*0.9/(1+2*x[i-1])*y[i-1]
print(y)
y_xn=init_y(x)
print(y_xn)
print(error(y))
```

```python
plt.figure
plt.plot(x, y_xn, 'r', label = 'Original curve')
plt.scatter(x,y)
plt.legend()
plt.show()
```

## 改进型Euler法
$$
y_{n+1} = y_n +\frac{h}{2}[f(x_n,y_n)+f(x_{n+1},y_n+hf(x_n,y_n))]

```python
x=np.linspace(0,0.1,6)
y = [0]*len(x)
y[0]=1
for i in range(1,len(y)):
    y[i] = y[i-1]+0.01*(-0.9/(1+2*x[i-1])*y[i-1]-0.9/(1+2*x[i])*(y[i-1]-0.02*0.9/(1+2*x[i-1])*y[i-1]))
print(y)
y_xn=init_y(x)
print(y_xn)
print(error(y))
```

```python
plt.figure
plt.plot(x, y_xn, 'r', label = 'Original curve')
plt.scatter(x,y)
plt.legend()
plt.show()
```
