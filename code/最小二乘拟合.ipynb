```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import leastsq 
import math
```

```python
def Fun(p, x):
    a0, a2 = p
    return a0  + a2 * x ** 2

def error(p, x, y):
    return Fun(p,x)-y

x = np.linspace(-2,2,5)
y = [0,1,2,1,0]
p0 = [100,0.1]
para = leastsq(error, p0, args=(x,y))
y_fitted = Fun(para[0], x)

plt.figure
plt.plot(x, y, 'r', label = 'Original curve')
plt.plot(x, y_fitted,'-b',label='Fitted curve')
plt.legend()
plt.show()
    
```

```python
def init_y(x):
    return x**2+x*np.cos(x)

def error(p, x, y):
    return Fun(p,x)-y

def Fun(p,x):
    res = 0
    for i in range (len(p)):
        res += p[i]*x**i
    return res

x = np.linspace(0,np.pi,1000)
y = init_y(x)
p0 = [100,0.1,0.1]

para = leastsq(error, p0, args=(x,y))
y_fitted = Fun(para[0], x)

plt.figure
plt.plot(x, y, 'r', label = 'Original curve')
plt.plot(x, y_fitted,'-b',label='Fitted curve')
plt.legend()
plt.show()
print(para[0])
```
